```python
# Softmax for reducing numerical roundoff error

import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

from sklearn.metrics import accuracy_score

model = Sequential([
    Dense(units=25, activation='relu'),
    Dense(units=15, activation='relu'),
    Dense(units=10, activation='linear')    # linear activation is used instead of softmax
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)  # the from_logits=True argument will calculate z instead of probability
)

path = '/kaggle/input/handwritten-digits/mnist.npz'
with np.load(path, allow_pickle=True) as f:
    x_train, y_train = f["x_train"], f["y_train"]
    x_test, y_test = f["x_test"], f["y_test"]
x_train = x_train.reshape(-1, 784)
x_test = x_test.reshape(-1, 784)

# Normalize the pixel values to range [0, 1]
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
x_train=x_train[:5000]
y_train=y_train[:5000]
model.fit(x_train, y_train, epochs=1000, verbose=0)

y_pred = model.predict(x_test)  #This y_pred will store the all the z values and not the probability

a=tf.nn.sigmoid(y_pred) #this will calculate the sigmoid value of z and the largest probability index will corresponds to the predicted number
```

```python
accuracy=accuracy_score(y_test, np.argmax(a, axis=1))
print(accuracy)
```
